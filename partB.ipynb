```python
vt_d = {
    0:[-0.015997038119123622,-0.01620390528828625,-0.016090023884358703],
    1:[-0.034949725553075746,-0.03559775118037491,-0.035145344610740474],
    2:[-0.053905315348262754,-0.05396064644575044,-0.05400442330617589],
    3:[-0.07252692065395268,-0.07254667797183476,-0.07234910715211447],
    4:[-0.09012628543032661,-0.09023536767910868],
    5:[-0.10683127777543794,-0.1068788367533117]
    # 6:[]
}
```

```python
m = []
vt = []
def mass(m):
    return (45.9 + 51.2*m)/1000
for k, v in enumerate(vt_d):
    m.append(float(k))
    vt.append(sum(vt_d[k])/len(vt_d[k]))
m = mass(np.array(m))
vt = np.array(vt)
```

```python
plt.scatter(m,vt,label='average terminal velocity')
m1,c1 = np.polyfit(m,vt,1)
# for i in range(len(m)):
plt.plot(m,m1*m+c1,label='straight line fit',color='orange')
plt.xlabel("mass (kg)")
plt.ylabel("v_t    (m/s)")
plt.legend()
plt.show()
print(m1,c1)
```

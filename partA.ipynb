```python
def mass(m):
    return (45.9 + 51.2*m)/1000

m = [1,2,2,3,5,6]
m1 = np.array([mass(i) for i in m])
a = [-0.006657492322521344,-0.01116744422424312,-0.011211353137250017,-0.015526301985440357,-0.024663137421091656,-0.029246537350135642]
```

```python
plt.scatter(m1,abs(np.array(a)),label='data')
slope,intercept = np.polyfit(m1,abs(np.array(a)),1)
plt.plot(m1,m1*slope+intercept,color='orange',label='fit')
plt.xlabel("mass (kg)")
plt.ylabel("a (m/s^2)")
plt.legend()
plt.show()
print(slope, intercept)
```

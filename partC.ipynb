```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
calibration = pd.read_csv("./calibration.csv")
```

```python
plt.scatter(calibration['d'],abs(calibration['B1']),s=2,label='data')
a,b,c = np.polyfit(calibration['d'],abs(calibration['B1']),2)
a1,b1,c1,d1 = np.polyfit(calibration['d'],abs(calibration['B1']),3)
# plt.plot(calibration['d'],a*calibration['d']**2+b*calibration['d']+c,color='orange',label='quadratic fit')
plt.plot(calibration['d'],a1*calibration['d']**3+b1*calibration['d']**2+c1*calibration['d']+d1,color='orange',label='cubic fit')
plt.legend()
plt.xlabel("d (mm)")
plt.ylabel("B (Gauss)")
plt.show()
print(a,b,c)
```

```python
df = pd.read_csv("./data/partC/s7t1.csv")
skipindex = 1000
df = df.iloc[:skipindex]

k = 300
m1,c1 = np.polyfit(df['t'][-k:],df['y'][-k:],1)
plt.scatter(df['t'],df['y'],s=1)
plt.plot(df['t'][-k:],m1*df['t'][-k:]+c1,label='fit',color='orange')
plt.xlabel("time (s)")
plt.ylabel("y (m)")
# plt.legend()
plt.show()
print(m1)
```

```python
dst = {
    0:17,
    1:18,
    2:19,
    3:20,
    4:21,
    5:22,
    7:24
}

def bd(distance):
    return a1*distance**3 + b1*distance**2 + c1*distance + d1
```
